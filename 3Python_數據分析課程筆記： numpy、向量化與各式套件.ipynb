# 3.7 從Array篩選出所需資料
* L = np.array([3, -2 , 4, -5, 333,-555, -60]) 

    L<0 會顯示出array([False, True, False, True, False, True, True])

* 使用array = array[條件]，取出符合條件的array

```python
L=np.array([3, -2 , 4, -5, 333,-555, -60]) 
```

```python
L[L<0]
```

* 範例從函數擷取需要資訊

```python
x = np.linspace(-5, 5 ,500)
y = np.sinc(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.plot(x[y>0], y[y>0], "o")
```
